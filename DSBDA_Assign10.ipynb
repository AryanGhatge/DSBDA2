```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv(url, names = column_names)
```

```python
print("Features and their types:")
print(iris_df.dtypes)
```

```python
iris_df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Features in Iris Dataset")
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris_df)
plt.title("Boxplots of Features in Iris Dataset")
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='class', y=feature, data=iris_df)
    plt.title(f"Boxplot of {feature}")
plt.tight_layout()
```
